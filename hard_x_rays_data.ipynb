```python
from astropy.io import fits
import matplotlib.pyplot as plt

hdul = fits.open('/content/lightcurve_czt1.fits')

bands = [
    "20-40 keV",
    "40-60 keV",
    "60-80 keV",
    "80-150 keV",
    "18-160 keV"
]

plt.figure(figsize=(15,8))

for i in range(1,6):
    d = hdul[i].data
    plt.plot(d['CTR'], label=bands[i-1], alpha=0.8)

plt.xlabel("Time Index")
plt.ylabel("Counts (CTR)")
plt.title("HEL1OS CZT2 Multi-Band Light Curves")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

for i in range(1,6):

    d = hdul[i].data

    peak_idx = np.argmax(d['CTR'])

    print(
        f"Band {i}",
        "Peak Time =", d['ISOT'][peak_idx],
        "Peak CTR =", d['CTR'][peak_idx]
    )
```
